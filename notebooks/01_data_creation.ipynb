```python
import pandas as pd
import random

# Set seed for reproducibility
random.seed(42)

# Define possible courses
courses = [
    "Python for Data Science",
    "Machine Learning Fundamentals",
    "Web Development with HTML CSS JavaScript",
    "SQL and Database Management",
    "Java Programming Basics",
    "Data Structures and Algorithms in Python",
    "Excel for Data Analysis",
    "Introduction to Artificial Intelligence",
    "React and Frontend Development",
    "Cloud Computing with AWS",
    "Statistics for Data Science",
    "Deep Learning with TensorFlow",
    "Mobile App Development with Flutter",
    "Cybersecurity Essentials",
    "Business Analytics with Tableau"
]

# Generate 30 student records
student_data = []

for i in range(1, 31):
    # Each student has taken 2-4 courses
    num_courses = random.randint(2, 4)
    selected_courses = random.sample(courses, num_courses)
    
    for course in selected_courses:
        record = {
            'student_id': f'S{i:03d}',  # S001, S002, etc.
            'name': f'Student{i}',
            'course_name': course,
            'completion_%': random.randint(60, 100),
            'grade': random.choice(['A+', 'A', 'A', 'B+', 'B', 'B']),
            'time_spent_hours': random.randint(20, 80)
        }
        student_data.append(record)

# Create DataFrame
df_students = pd.DataFrame(student_data)

# Display first 10 rows
print("Student Data Created!")
print(f"Total records: {len(df_students)}")
print("\nFirst 10 rows:")
df_students.head(10)
```

```python
def extract_skills_from_text(text, skill_list):
    """
    Extract skills from a given text by matching against skill list
    
    Parameters:
    - text: string (course name or job description)
    - skill_list: list of skills to match
    
    Returns:
    - list of found skills
    """
    found_skills = []
    text_lower = text.lower()
    
    for skill in skill_list:
        # Check if skill appears in text (case-insensitive)
        if skill.lower() in text_lower:
            found_skills.append(skill)
    
    return found_skills


# Test the function
test_course = "Python for Data Science and Machine Learning"
extracted = extract_skills_from_text(test_course, all_skills)
print(f"Course: {test_course}")
print(f"Extracted skills: {extracted}")

print("\n" + "="*60)

# Extract skills from all student courses
print("\nExtracting skills from student data...")
df_students['extracted_skills'] = df_students['course_name'].apply(
    lambda x: extract_skills_from_text(x, all_skills)
)

# Show examples
print("\nExamples:")
for i in range(5):
    print(f"\nCourse: {df_students.iloc[i]['course_name']}")
    print(f"Skills: {df_students.iloc[i]['extracted_skills']}")

# Save updated student data
df_students.to_csv('../data/student_data_with_skills.csv', index=False)
print("\n‚úÖ Skills extracted and saved!")
```

```python
def calculate_skill_proficiency(completion, grade, time_spent):
    """
    Calculate skill proficiency score
    
    Parameters:
    - completion: completion percentage (0-100)
    - grade: letter grade (A+, A, B+, etc.)
    - time_spent: hours spent on course
    
    Returns:
    - proficiency score (0-1)
    """
    # Grade mapping
    grade_map = {
        'A+': 1.0,
        'A': 0.9,
        'B+': 0.8,
        'B': 0.7,
        'C+': 0.6,
        'C': 0.5
    }
    
    grade_score = grade_map.get(grade, 0.5)
    
    # Normalize completion (0-1)
    completion_score = completion / 100
    
    # Normalize time (assume 50 hours is standard)
    time_score = min(time_spent / 50, 1.0)
    
    # Weighted average: completion (50%), grade (30%), time (20%)
    proficiency = (completion_score * 0.5) + (grade_score * 0.3) + (time_score * 0.2)
    
    return round(proficiency, 2)


# Test the function
test_result = calculate_skill_proficiency(85, 'A', 45)
print(f"Test: 85% completion, grade A, 45 hours")
print(f"Proficiency score: {test_result}")

print("\n" + "="*60)

# Calculate proficiency for all records
print("\nCalculating proficiency scores...")
df_students['proficiency_score'] = df_students.apply(
    lambda row: calculate_skill_proficiency(
        row['completion_%'], 
        row['grade'], 
        row['time_spent_hours']
    ), 
    axis=1
)

# Show examples
print("\nExamples:")
print(df_students[['student_id', 'course_name', 'completion_%', 'grade', 'proficiency_score']].head(10))
```

```python
def build_student_profile(student_id, df):
    """
    Build skill profile for a specific student
    
    Parameters:
    - student_id: student identifier
    - df: dataframe with student data
    
    Returns:
    - dictionary with skills and proficiency scores
    """
    # Filter data for this student
    student_courses = df[df['student_id'] == student_id]
    
    profile = {}
    
    for _, row in student_courses.iterrows():
        skills = row['extracted_skills']
        proficiency = row['proficiency_score']
        
        for skill in skills:
            # If skill already exists, take the maximum proficiency
            if skill in profile:
                profile[skill] = max(profile[skill], proficiency)
            else:
                profile[skill] = proficiency
    
    return profile


# Build profiles for all students
print("Building student profiles...\n")

student_profiles = {}

for student_id in df_students['student_id'].unique():
    profile = build_student_profile(student_id, df_students)
    student_profiles[student_id] = profile

print(f"‚úÖ Profiles created for {len(student_profiles)} students")

# Show first 3 student profiles
print("\nExample Profiles:")
for i, (student_id, profile) in enumerate(list(student_profiles.items())[:3]):
    print(f"\n{student_id}:")
    for skill, score in profile.items():
        print(f"  - {skill}: {score}")
```

```python
# %pip install scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def match_jobs_for_student(student_profile, jobs_df):
    """
    Find matching jobs for a student based on their skill profile
    
    Parameters:
    - student_profile: dict with skills and proficiency
    - jobs_df: dataframe with job information
    
    Returns:
    - dataframe with matched jobs sorted by match percentage
    """
    # Convert student profile to string of skills
    student_skills = ' '.join(student_profile.keys())
    
    # Get job skills
    job_skills = jobs_df['required_skills'].tolist()
    
    # Combine for vectorization
    all_texts = [student_skills] + job_skills
    
    # Create TF-IDF vectors
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(all_texts)
    
    # Calculate cosine similarity
    # Student is first row, jobs are rest
    similarities = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()
    
    # Create result dataframe
    result = jobs_df.copy()
    result['match_%'] = (similarities * 100).round(2)
    
    # Sort by match percentage
    result = result.sort_values('match_%', ascending=False)
    
    return result


# Test with first student
test_student_id = 'S001'
test_profile = student_profiles[test_student_id]

print(f"Testing job matching for {test_student_id}")
print(f"Student skills: {list(test_profile.keys())}\n")

matched_jobs = match_jobs_for_student(test_profile, df_jobs)

print("Top 5 Job Recommendations:")
print(matched_jobs[['job_title', 'company', 'required_skills', 'match_%']].head())
```

```python
def analyze_skill_gap(student_profile, job_required_skills):
    """
    Find missing skills for a job
    
    Parameters:
    - student_profile: dict with student skills
    - job_required_skills: string with comma-separated skills
    
    Returns:
    - list of missing skills
    """
    student_skills = set(student_profile.keys())
    
    # Parse job skills
    job_skills_list = [s.strip() for s in job_required_skills.split(',')]
    job_skills = set(job_skills_list)
    
    # Find gaps
    missing_skills = job_skills - student_skills
    matching_skills = student_skills & job_skills
    
    return {
        'missing': list(missing_skills),
        'matching': list(matching_skills),
        'match_count': len(matching_skills),
        'total_required': len(job_skills)
    }


# Demonstrate skill gap analysis
print("SKILL GAP ANALYSIS EXAMPLES")
print("="*60)

for student_id in list(student_profiles.keys())[:3]:
    print(f"\nüìä {student_id}")
    profile = student_profiles[student_id]
    recommendations = all_recommendations[student_id]
    
    # Analyze top 3 jobs
    for i in range(3):
        job = recommendations.iloc[i]
        gap_analysis = analyze_skill_gap(profile, job['required_skills'])
        
        print(f"\n  Job: {job['job_title']} ({job['match_%']}% match)")
        print(f"  ‚úÖ You have: {gap_analysis['matching']}")
        print(f"  ‚ùå You need: {gap_analysis['missing']}")
        print(f"  Progress: {gap_analysis['match_count']}/{gap_analysis['total_required']} skills")
```
